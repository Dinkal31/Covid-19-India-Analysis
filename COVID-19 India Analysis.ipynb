Plotting the pie chart for different age group affected with COVID-19 in india

```python
def plot_pie_charts(x, y, title):
    c = random.choices(list(mcolors.CSS4_COLORS.values()),k = 10)
    plt.figure(figsize=(15,10))
    plt.title(title, size=15)
    plt.pie(y, colors = c)
    plt.legend(x, loc='best', fontsize=15)
    plt.show()
```

```python
#Creating a function to plot pie charts and plotting a pie chart of the age distribution of the confirmed cases.
plot_pie_charts(df1['AgeGroup'], df1['TotalCases'], 'Age distribution')
```

```python
def plot_bar_graphs(x, y, title):
    plt.figure(figsize=(20, 12))
    plt.barh(x, y)
    plt.title(title, size=20)
    plt.xticks(size=15)
    plt.yticks(size=20)
    plt.show()
df2 = df2.drop(df2.index[32])
```

Plotting the grpah between states and number of rural hospitals in those states.

```python
plot_bar_graphs(df2['State/UT'], df2['NumRuralHospitals_NHP18'], 'No of Rural Hospitals in different regions')
```

Plotting the grpah between states and number of rural beds in those states.

```python
plot_bar_graphs(df2['State/UT'], df2['NumRuralBeds_NHP18'], 'No of Rural Beds in different regions')
```

Plotting the grpah between states and number of Urban Hospitals in those states.

```python
plot_bar_graphs(df2['State/UT'], df2['NumUrbanHospitals_NHP18'], 'No of Urban Hospitals in different regions')
```

Plotting the grpah between states and number of Urban Beds in those states.
